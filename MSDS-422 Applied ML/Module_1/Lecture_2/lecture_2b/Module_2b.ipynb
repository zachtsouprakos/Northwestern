```python
df = pd.read_csv(INFILE)
dt = df.dtypes
```

```python
objList = []
numList = []

for i in dt.index:
    if i in ([TARGET_F, TARGET_A]) : continue
    if dt[i] in (['object']) : objList.append(i)
    if dt[i] in (['float64','int64']) : numList.append(i)
```
